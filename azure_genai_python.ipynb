```python
import os
import requests
import json

# Load from environment
api_key = os.environ.get("AZURE_OPENAI_KEY", "").strip('"')
endpoint = os.environ.get("AZURE_OPENAI_ENDPOINT", "").strip('"')

# Fix - use the endpoint directly as the URL (don't add anything to it)
headers = {
    "Content-Type": "application/json",
    "api-key": api_key
}

body = {
    "messages": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is machine learning in 2 sentences?"}
    ]
}

# Use endpoint directly as the full URL
response = requests.post(endpoint, headers=headers, json=body)
result = response.json()
print(result["choices"][0]["message"]["content"])
```
